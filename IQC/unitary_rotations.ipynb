We shall represent a state as
$$
\begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}=
\begin{bmatrix}
a & b \\ x & y
\end{bmatrix}
$$

Then
$$
R_x(2\theta) \begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}=
\begin{pmatrix}
\cos\theta & -i\sin\theta \\
-i\sin\theta &\cos\theta
\end{pmatrix}
\begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}
=
\begin{pmatrix}
\alpha\cos\theta-i\chi\sin\theta \\ \chi\cos\theta-i\alpha\sin\theta
\end{pmatrix}
$$

$$
R_x(2\theta) \begin{bmatrix}
a & b \\ x & y
\end{bmatrix}=
\begin{pmatrix}
\cos\theta & -i\sin\theta \\
-i\sin\theta &\cos\theta
\end{pmatrix}
\begin{bmatrix}
a & b \\ x & y
\end{bmatrix}
=
\begin{bmatrix}
a\cos\theta+y\sin\theta & b\cos\theta-x\sin\theta \\ x\cos\theta+b\sin\theta & y\cos\theta-a\sin\theta
\end{bmatrix}
$$

We see that each dimension of $\alpha,\chi$ is interacting with the cross dimension of their counterpart.

Then
$$
R_y(2\theta) \begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}=
\begin{pmatrix}
\cos\theta & -\sin\theta \\
\sin\theta &\cos\theta
\end{pmatrix}
\begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}
=
\begin{pmatrix}
\alpha\cos\theta-\chi\sin\theta \\ \chi\cos\theta+\alpha\sin\theta
\end{pmatrix}
$$

$$
R_y(2\theta) \begin{bmatrix}
a & b \\ x & y
\end{bmatrix}=
\begin{pmatrix}
\cos\theta & -\sin\theta \\
\sin\theta &\cos\theta
\end{pmatrix}
\begin{bmatrix}
a & b \\ x & y
\end{bmatrix}
=
\begin{bmatrix}
a\cos\theta-x\sin\theta & b\cos\theta-y\sin\theta \\
x\cos\theta+a\sin\theta & y\cos\theta+b\sin\theta
\end{bmatrix}
$$

We see that each dimension of $\alpha,\chi$ is interacting with the same dimension of their counterpart.

Then
$$
R_z(2\theta) \begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}=
\begin{pmatrix}
\cos\theta - i\sin\theta & 0 \\
0 & \cos\theta + i\sin\theta
\end{pmatrix}
\begin{pmatrix}
\alpha \\ \chi
\end{pmatrix}
=
\begin{pmatrix}
\alpha\cos\theta - i\alpha\sin\theta \\ \chi\cos\theta+i\chi\sin\theta
\end{pmatrix}
$$

$$
R_z(2\theta) \begin{bmatrix}
a & b \\ x & y
\end{bmatrix}=
\begin{pmatrix}
\cos\theta - i\sin\theta & 0 \\
0 & \cos\theta + i\sin\theta
\end{pmatrix}
\begin{bmatrix}
a & b \\ x & y
\end{bmatrix}
=
\begin{bmatrix}
a\cos\theta+b\sin\theta & b\cos\theta-a\sin\theta \\
x\cos\theta-y\sin\theta & y\cos\theta+x\sin\theta
\end{bmatrix}
$$

We see that each dimension of $\alpha,\chi$ is interacting with their own cross dimension.

We typically use the H gate to introduce a superposition from pure states, for ease of representation and involution (and related). $\ket{0}$ and $\ket{1}$ get mapped to superpositions in real coefficients. $R_Y(\pi/2)$ is similar, but $\ket{0}$ gets mapped to a negative coefficient.
