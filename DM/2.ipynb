```python
def bhattacharyya_coefficient(p, q):
    """Compute the Bhattacharyya coefficient for two discrete 
       probability distributions.
    
    Args:
        p (np.array[float]): A probability distribution vector.
        q (np.array[float]): A probability distribution vector.
        
    Returns:
        (float): A number that represents the Bhattacharyya coefficient
        between two probability distribution vectors.
    """

    ##################
    # YOUR CODE HERE #
    ##################


    # RETURN THE BHATTACHARYYA COEFFICIENT
    return np.dot(np.sqrt(p), np.sqrt(q))
```

## D.2.2

```python
def kolmogorov_distance(p, q):
    """Compute the Kolmogorov distance for two discrete 
       probability distributions.
    
    Args:
        p (np.array[float]): A vector of probability distribution
        q (np.array[float]): A vector of probability distribution.
        
    Returns:
        (float): A number that represents the trace distance
        of the probability distributions.
    """
    return np.sum(np.abs(p-q))/2


def kolmogorov_lower_bound(p, q):
    """Compute the lower bound for the trace distance
       between two discrete probability distributions using
       the Bhattacharyya coefficient.
    
    Args:
        p (np.array[float]): A probability distribution vector.
        q (np.array[float]): A probability distribution vector.
        
    Returns:
        (float): A number that represents the lower bound for
        the trace distance between two probability
        distribution vectors.
    """

    ##################
    # YOUR CODE HERE #
    ##################


    # RETURN THE LOWER BOUND USING BHATTACHARYYA INEQUALITY
    return 1 - bhattacharyya_coefficient(p,q)


def kolmogorov_upper_bound(p, q):
    """Compute the upper bound for the trace distance
       between two discrete probability distributions using
       the Bhattacharyya coefficient.
    
    Args:
        p (np.array[float]): A probability distribution vector.
        q (np.array[float]): A probability distribution vector.
        
    Returns:
        (float): A number that represents the upper bound for
        the trace distance between two probability
        distribution vectors.
    """

    ##################
    # YOUR CODE HERE #
    ##################


    # RETURN THE UPPER BOUND USING BHATTACHARYYA INEQUALITY
    return np.sqrt(1 - bhattacharyya_coefficient(p,q) ** 2)
    

P = np.array([0.5,0.5])
Q = np.array([0.25,0.75])

print("The trace distance for P and Q is: " , kolmogorov_distance(P,Q))
print("Lower bound: ", kolmogorov_lower_bound(P,Q))
print("Upper bound: ",kolmogorov_upper_bound(P,Q))
```
